# Cancer Classification 

```python
import numpy as np
import torch 
from torch import nn
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv("Cancer_Data.csv")
df.head()
```

```python
df.info()
```

```python
df.dropna(axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["diagnosis"] = le.fit_transform(df["diagnosis"])
```

```python
df.head()
```

```python
col_x = ["radius_mean",'texture_mean','smoothness_mean','concavity_mean','symmetry_mean','fractal_dimension_mean']
col_y = ['diagnosis']
X = df[col_x]
y = df[col_y]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,random_state=5)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
se = StandardScaler()
X_train = se.fit_transform(X_train)
X_test = se.transform(X_test)
```

```python
X_test
```

```python
class LogisticRegression(nn.Module):
    def __init__(self,n_in):
        super().__init__()
        self.in_layer = nn.Linear(in_features=n_in,out_features=1)
        
    def forward(self,x: torch.Tensor) -> torch.Tensor:
        return self.in_layer(x)
```

```python
model = LogisticRegression(6)
loss_func = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(params=model.parameters(),lr=0.1)
```

```python
X_train = torch.from_numpy(X_train).type(torch.float)
y_train = torch.from_numpy(y_train.values).type(torch.float)
X_test = torch.from_numpy(X_test).type(torch.float)
y_test = torch.from_numpy(y_test.values).type(torch.float)
```

```python
def accuracy_fn(y_true, y_pred):
    correct = torch.eq(y_true, y_pred).sum().item() # torch.eq() calculates where two tensors are equal
    acc = (correct / len(y_pred)) * 100 
    return acc
```

```python
ep = 500

for i in range(ep):
    model.train()
    yloss = torch.sigmoid(model(X_train).squeeze())
    y_pred = torch.round(yloss)
    loss = loss_func(yloss,y_train.squeeze())
    acc = accuracy_fn(y_train.squeeze(),y_pred)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    model.eval()
    with torch.inference_mode():
        yloss = torch.sigmoid(model(X_test).squeeze())
        yp = torch.round(torch.sigmoid(model(X_test).squeeze()))
        loss_test = loss_func(yloss,y_test.squeeze())
        acc_test = accuracy_fn(y_test.squeeze(),yp)
        
    if i%10 == 0:
        print(f"Epoch: {i} | Train Loss: {loss} | Train Accuracy: {acc} | Test Loss: {loss_test} | Test Accuracy: {acc_test}")
    
```
